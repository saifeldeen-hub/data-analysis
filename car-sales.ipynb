```python
df['body'].value_counts()
```

```python
body_count=df.groupby('body')['body'].value_counts().reset_index()
body_count
```

```python
x=len(body_count)
x
```

```python
top=body_count.head(30)
```

```python
for i in range (0,len(top),10):
    plt.figure(figsize=(20,10))
    plt.title('body count in dataframe')
    plt.xlabel('body')
    plt.ylabel('count')
    x=sns.barplot(x=top.iloc[i:i+10]['body'],y=top.iloc[i:i+10]['count'],data=top)
    for container in x.containers:
        x.bar_label(container)
    plt.show()
```
